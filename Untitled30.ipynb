```python
# 2 ans)
import pandas as pd

# Assuming you have the DataFrame 'df' with columns 'A', 'B', and 'C'
# For example:
data = {
    'A': [10, 20, 30, 40],
    'B': [50, 60, 70, 80],
    'C': [90, 100, 110, 120]
}

df = pd.DataFrame(data)
def reindex_dataframe(df):
    # Create a new index starting from 1 and incrementing by 2
    new_index = pd.RangeIndex(start=1, stop=len(df) * 2 + 1, step=2)
    
    # Assign the new index to the DataFrame
    df = df.set_index(new_index)
    
    return df
# Call the function to reindex the DataFrame
df = reindex_dataframe(df)

# The DataFrame will now have a new index starting from 1 and incrementing by 2
print(df)
```

```python
import pandas as pd

def count_words(df):
    def word_count(text):
        words = text.split()  # Split the text by whitespace to get individual words
        return len(words)     # Return the count of words

    # Apply the custom function to the 'Text' column and create a new column 'Word_Count'
    df['Word_Count'] = df['text'].apply(word_count)

data = {
    'text': [
        'this is a sample text',
        'pandas DataFrame is awesome',
        'hello, how are you?'
    ]
}

df = pd.DataFrame(data)
count_words(df)  # Call the function without reassigning the DataFrame
print(df)
```

```python

"""DataFrame.size returns the total number of elements in the DataFrame, which is equal to the number of rows multiplied by the number of columns.
The result is a single integer value representing the size of the DataFrame.
DataFrame.shape():

DataFrame.shape returns a tuple representing the dimensions of the DataFrame.
The result is a tuple of two elements: the first element is the number of rows, and the second element is the number of columns."""

import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'San Francisco', 'Los Angeles']
}

df = pd.DataFrame(data)

# DataFrame.size()
print("DataFrame.size:", df.size)
# Output: DataFrame.size: 9 (3 rows x 3 columns = 9 elements)

# DataFrame.shape()
print("DataFrame.shape:", df.shape)
# Output: DataFrame.shape: (3, 3) (3 rows, 3 columns)
```

```python
#ans7)
import pandas as pd

def extract_username(df):
    # Split the 'Email' column on '@' and extract the first part (username)
    df['Username'] = df['Email'].str.split('@').str[0]
    return df

# Example usage:
data = {
    'Email': ['user1@example.com', 'user2@example.com', 'user3@example.com']
}

df = pd.DataFrame(data)
df = extract_username(df)

print(df)
```

```python
#ans8)
import pandas as pd

def filter_dataframe(df):
    # Boolean indexing to select rows that meet the conditions
    selected_rows = df[(df['A'] > 5) & (df['B'] < 10)]
    return selected_rows

# Example usage:
data = {
    'A': [3, 7, 2, 8, 6],
    'B': [15, 5, 9, 3, 12],
    'C': [10, 20, 30, 40, 50]
}

df = pd.DataFrame(data)
selected_df = filter_dataframe(df)

print(selected_df)
```

```python
#ans9)
import pandas as pd

def calculate(df):
    mean_value= df['values'].mean()
    median_value=df['values'].median()
    std_value=df['values'].std()
    
    return  mean_value,median_value,std_value

    
    
    
data = {
    "values": [1,2,3,4,5]
}
df=pd.DataFrame(data)
mean,median,std=calculate(df)
print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std)
    
```

```python
#ans10)
import pandas as pd

def calculate_moving_average(df):
    # Sort the DataFrame by the 'Date' column (if it's not already sorted)
    df = df.sort_values(by='Date')
    
    # Calculate the moving average using a window of size 7 and including the current day
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()
    
    return df

# Example usage:
data = {
    'Date': pd.date_range(start='2023-07-01', periods=10, freq='D'),
    'Sales': [100, 120, 80, 90, 110, 150, 130, 95, 105, 140]
}

df = pd.DataFrame(data)
df = calculate_moving_average(df)

print(df)
```

```python
#ans11)
import pandas as pd
from datetime import datetime

def add_weekday_column(df, date_column='Date', weekday_column='Weekday'):
    # Convert the 'Date' column to a datetime data type if it's not already
    df[date_column] = pd.to_datetime(df[date_column])
    
    # Extract the weekday name for each date
    df[weekday_column] = df[date_column].dt.strftime('%A')
    
    # Drop the other column if it exists
    df.drop(columns=['OtherColumn'], inplace=True, errors='ignore')
    
    return df

# Example usage without sample data
# Assuming you have a DataFrame named 'df' with a column 'Date' and 'OtherColumn'
# Call the function to add the 'Weekday' column and remove the 'OtherColumn'
df = add_weekday_column(df)
df
# Now your 'df' DataFrame will have only the 'Date' and 'Weekday' columns.
```

```python
#ans12)
import pandas as pd

def select_rows_between_dates(df, date_column='Date', start_date='2023-01-01', end_date='2023-01-31'):
    # Convert the 'Date' column to a datetime data type if it's not already
    df[date_column] = pd.to_datetime(df[date_column])
    
    # Filter rows based on the given date range
    selected_rows = df[(df[date_column] >= start_date) & (df[date_column] <= end_date)]
    
    return selected_rows

# Example usage:
# Assuming 'df' is your Pandas DataFrame with a column 'Date'
# and you want to select rows between '2023-01-01' and '2023-01-31'
selected_rows = select_rows_between_dates(df)
df
# Now 'selected_rows' will contain all rows where the 'Date' is between '2023-01-01' and '2023-01-31'.
```
